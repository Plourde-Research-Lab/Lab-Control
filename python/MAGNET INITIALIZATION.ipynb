```python
import pyvisa
import numpy as np
import time
```

# Connect to Magnet Power Supply

```python
rm = pyvisa.ResourceManager()
rm.list_resources()
```

**Make GPIB Connection**

```python
ps = rm.open_resource('GPIB0::3::INSTR')
ps.ask('*IDN?')
```

**Enable Output and set current limit**

```python
ps.write('VOLT:LEV:IMM 0V')
ps.write('CURRENT:LEV:IMM 9.2A')
ps.write('OUTPUT 1')
```

```python
ps.ask("Volt?")
```

##Magup/Magdown here

```python
start = float(ps.ask('VOLT?'))*1000
stop = 0  #Change to 390 to Magup, 0 to Magdown <<------------------------------- Control here
for i in np.linspace(start,stop, np.abs(stop-start)+1):
    ps.write('VOLT:LEV:IMM ' + str(i) + 'mV')
    print(str(i) + "mV\n" + "Current: " + ps.ask('MEAS:CURR?') + "Voltage: " + ps.ask('MEAS:VOLT?'))
    time.sleep(5)
```

If there's an error here, just re-run the cell. It automatically finds the last value it was programmed at.
